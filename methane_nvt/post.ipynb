```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import mdtraj as mdt

traj = mdt.load('nvt.trr', top='nvt.gro')
pairs = traj.top.select_pairs("name =~ '_CH4'", "name =~ '_CH4'")
rdf = mdt.compute_rdf(traj, pairs, bin_width=0.005)
traj.n_frames
```

```python
plt.plot(rdf[0],rdf[1])
```

```python
energy = np.loadtxt('energy.xvg',skiprows=24)

plt.plot(energy[:,0], energy[:,1])
plt.axhline(y=np.mean(energy[:,1]), color="k", linestyle="--", linewidth=1.5)
plt.xlabel('Time (ps)')
plt.ylabel('Energy (kJ/mol)')
```

```python
temperature = np.loadtxt('temperature.xvg',skiprows=24)

plt.plot(temperature[:,0], temperature[:,1])
plt.axhline(y=np.mean(temperature[:,1]), color="k", linestyle="--", linewidth=1.5)
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
```
